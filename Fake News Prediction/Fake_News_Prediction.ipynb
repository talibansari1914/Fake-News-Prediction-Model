```python
# Important imports 
import pandas as pd
import numpy as np
import re 
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Loading the Data & Understanding 

```python
# Data Loading
news_dataset=pd.read_csv(r"C:\Users\talib\Downloads\train.csv")
```

```python
# finding the Null Values
news_dataset.isnull().sum()
```

```python
# Filling the null Values 
news_dataset=news_dataset.fillna('')
```

```python
# Merging 2 columns Author + Title
# Reason : Because it will give better result & text data will be more complex
news_dataset["content"]=news_dataset["author"] + " " + news_dataset["title"]
news_dataset["content"].head()
```

```python
# Seprating the Data 
X_text=news_dataset.drop(columns="label",axis=1)
y=news_dataset["label"]

print(X_text)
print(y)
```

### Stemming
**stemming is the process of reducing word to it's root word<br>
example: actor , acting , action -----> act (root word)**

```python
port_stem = PorterStemmer()
```

```python
def stemming(content):
    stemmed_content = re.sub("[^a-zA-Z]"," ",content) # taking only english character 
    stemmed_content = stemmed_content.lower() # lower case all the data
    stemmed_content = stemmed_content.split() # split the data into the list
    stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words("english")]
    stemmed_content = " ".join(stemmed_content)
    return stemmed_content
```

```python
news_dataset["content"] = news_dataset["content"].apply(stemming)
```

```python
print(news_dataset["content"])
```

```python
# Seprating the Data
X=news_dataset["content"].values
y=news_dataset["label"].values
print(y)
```

```python
print(X)
```

### Converting Textual Data to Numerical


```python
# Transform data from text to numerica 
# Reason : Machine does'nt understand text it only understand numeric values
vectorizer=TfidfVectorizer()
vectorizer.fit(X)

X=vectorizer.transform(X)
```

```python
print(X)
```

# Model Trainig and data splitting

```python
# data splitting
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=42)

# Model object 

lr=LogisticRegression()

# Training 
lr.fit(X_train,y_train)
```

# Accuracy & Prediction ---> Model Evaluation

```python
# prediction & accuracy on trainig data
X_train_predict=lr.predict(X_train)
accuracy_train=accuracy_score(y_train,X_train_predict)
print(f"Accuracy of the Model on Training Data:{accuracy_train:.4f}")
```

```python
# prediction & accuracy on testing data
X_test_predict=lr.predict(X_test)
accuracy_test=accuracy_score(y_test,X_test_predict)
print(f"Accuracy of the Model on Training Data:{accuracy_test:.4f}")
```
